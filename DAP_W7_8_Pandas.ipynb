# Lập lại chỉ mục (Reindexing)

Thay đổi labels của hàng hoặc cột. Các phép toán có thể:

+ Thay đổi thứ tự của dữ liệu để phù hợp với label mới
+ Thêm giá trị còn thiếu vào vị trí của label


```python
import numpy as np

N=20

df = pd.DataFrame({
   'A': pd.date_range(start='2016-01-01', periods=N, freq='D'),
   'x': np.linspace(0,stop=N-1,num=N),
   'y': np.random.rand(N),
   'C': np.random.choice(['Low','Medium','High'], N).tolist(),
   'D': np.random.normal(100, 10, size=(N)).tolist()
})
print(df)
```

## Filling while reindexing
Hàm reindex() có tham số filling với các giá trị sau:

+ pad/ffill: Điền các giá trị tiếp theo
+ bfill/backfill: Điền các giá trị trước đó
+ nearest: Điền các giá trị gần index nhất

```python
df1 = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df2 = pd.DataFrame(np.random.randn(2,3),columns=['col1','col2','col3'])
print(df1)
print(df2)
```

```python
# Padding NAN's
print(df2.reindex_like(df1))
```

```python
# Now Fill the NAN's with preceding Values
print("Data Frame with Forward Fill:")
print(df2.reindex_like(df1, method='ffill'))
```

```python
print(df2.reindex_like(df1, method='bfill'))
```

```python
print(df2.reindex_like(df1, method='nearest'))
```

```python
## Thiết lập limit số hàng được fill
print(df2.reindex_like(df1, method='ffill', limit=2))
```

# apply # transform

(1): Giống nhau

- transform() works with function, a string function, a list of functions, and a dict. However, apply() is only allowed with function.


(2): Khác nhau

- transform() cannot produce aggregated results.

- apply() works with multiple Series at a time. But, transform() is only allowed to work with a single Series at a time.

```python
## Giống nhau

print(df.transform('sqrt'))
print(df.transform([np.sqrt, np.exp]))
print(df.transform({
    'P': np.sqrt,
    'Q': np.exp,
}))
```

```python
df.apply({
    'P': np.sqrt,
    'Q': np.exp,
})
```

```python
## Khác nhau

df.apply(lambda x:x.sum())
```

```python
df.transform(lambda x:x.sum())
```

```python
# It is working
df.apply(lambda x: x['Q'] - x['P'], axis=1)
```

```python
# Getting error
df.transform(lambda x: x['Q'] - x['P'], axis=1)
```

## Kết hợp với groupby()

Cả 2 apply() và transform() đều kết hợp được với groupby(). Tuy nhiên:

- transform() returns a DataFrame that has the same length as the input

- apply() works with multiple Series at a time. But, transform() is only allowed to work with a single Series at a time.

```python
df = pd.DataFrame({
    'key': ['a','b','c'] * 4,
    'A': np.arange(12),
    'B': [1,2,3] * 4,
})
df
```

```python
## (1) transform() returns a Series that has the same length as the input

# Aggregating the sum of the given Series
def group_sum(x):
    return x.sum()
```

```python
gr_data_ap = df.groupby('key')['A'].apply(group_sum)
gr_data_ap
```

```python
gr_data_tr = df.groupby('key')['A'].transform(group_sum)
gr_data_tr
```

```python
## (2) apply() works with multiple Series at a time. But transform() is only allowed to work with a single Series at a time.

def subtract_two(x):
    return x['B'] - x['A']
```

```python
df.groupby('key').apply(subtract_two)
```

```python
df.groupby('key').transform(subtract_two)
```

# groupby()

## Tổng hợp: agg()

Hàm tổng hợp trả lại giá trị tổng hợp của mỗi group.

- It is defined as a function that returns a single aggregated value for each of the groups. We can perform several aggregation operations on the grouped data when the groupby object is created.


```python
import numpy as np

grouped = df.groupby('Year')
grouped['Points'].agg(np.mean)
```

```python
## Dùng hàm size để xem kích thước mỗi nhóm

groups = df.groupby('Team')
groups.agg(np.size)
```

### Tổng hợp nhiều hàm cùng lúc

Có thể truyền một list hoặc dict functions để thực thi tổng hợp.

## Clean empty cells

- Nếu data lớn có thể xóa empty cells bằng cách remove cả hàng.
- Nếu data nhỏ thì nên thay thế empty cells bằng giá trị khác.

```python
data.dropna(inplace=False)   # True: thực thi trên data
```

```python
data.fillna(130, inplace=False)         # inplace=True: thực thi trên original data
```

- Thực thi cho một column nhất định bằng cách chọn column đó.

```python
data["Calories"].fillna(130, inplace = True)
```

- Có thể thay thế bằng các giá trị như: Mean, Median, Mode

```python
x1 = data["Calories"].mean()
x2 = data["Calories"].median()
x3 = data["Calories"].mode()

data["Calories"].fillna(x1, inplace = True)
```

## Clean wrong format

```python
data['Date'] = pd.to_datetime(data['Date'])

print(data.to_string())
```

```python
data.dropna(subset=['Date'], inplace = True)
data
```

## Clean wrong data


```python
data.loc[7, 'Duration'] = 45
data
```

```python
for x in data.index:
  if data.loc[x, "Duration"] > 120:
    data.loc[x, "Duration"] = 120
```

```python
for x in data.index:
  if data.loc[x, "Duration"] > 120:
    data.drop(x, inplace = True)
```

## Clean duplicated


```python
print(data.duplicated())
```

```python
## Remove duplicates

data.drop_duplicates(inplace=True)
data
```
